```python
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
def plot_curve(file):
    res = pkl.load(open(file, 'rb'))
    x, y = list(zip(*res))
    plt.plot(x, y)
    plt.show()
```

```python
plot_curve('res_4_actions_0.pkl')
```

```python
plot_curve('res_4_actions_1.pkl')
```

```python
plot_curve('res_8_actions_0.pkl')
```
